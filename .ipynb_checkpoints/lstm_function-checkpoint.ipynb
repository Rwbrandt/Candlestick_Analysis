```python
import pandas as pd
```

```python
# PRIMARY STARTING DATAFRAME
ticker = "MSFT"
inter = "1m" #crypto has to be every 5 minutes while stocks is 1m

df = pd.read_csv(f"{ticker}_prices_{inter}.csv", parse_dates=True, infer_datetime_format=True)
df = df.set_index(pd.DatetimeIndex(df['Datetime']), drop=True, append=False, inplace=False, verify_integrity=False)
```

```python
# CLEANED/SIMPLIFIED DATAFRAME

# New dataframe for this specific technique.  Consists of OHLC only
ohlc_df = df[['Open', 'High', 'Low', 'Close']]


# slice dataframe to just last day
ohlc_df = ohlc_df.iloc[2340:2350,:]
ohlc_df
```

```python
# >>> LSTM FUNCTION <<<


def lstm_function(df, x_column_number, y_column_number, window_size, number_units, dropout_fraction):

    global predictions_df
    
    # Import necessary libraries
    import numpy as np
    import pandas as pd
    from sklearn.preprocessing import MinMaxScaler
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import LSTM, Dense, Dropout

    # This function chunks the data up with a rolling window of Xt - window to predict Xt.
    def window_data(df, window, x_column_number, y_column_number):
        X = []
        y = []
        for i in range(len(df) - window - 1):
            features = df.iloc[i:(i + window), x_column_number]
            target = df.iloc[(i + window), y_column_number]
            X.append(features)
            y.append(target)
        return np.array(X), np.array(y).reshape(-1, 1)
    
    # Creating the features (X) and target (y) data using the window_data() function
    X, y = window_data(df, window_size, x_column_number, y_column_number)
    
    # Splitting Data Between Training and Testing Sets
    split = int(0.7 * len(X))
    X_train = X[: split]
    X_test = X[split:]
    y_train = y[: split]
    y_test = y[split:]
    
    # Create a MinMaxScaler object
    scaler = MinMaxScaler()

    # Fit the MinMaxScaler object with the training feature data X_train
    scaler.fit(X_train)

    # Scale the features training and testing sets
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)

    # Fit the MinMaxScaler object with the training target data y_train
    scaler.fit(y_train)

    # Scale the target training and testing sets
    y_train = scaler.transform(y_train)
    y_test = scaler.transform(y_test)
    
    # Reshape Features Data for the LSTM Model
    X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
    X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
    
    # Define the LSTM RNN model
    model = Sequential()

    # activation = "sigmoid"
    
    # Layer 1
    model.add(LSTM(units=number_units, return_sequences=True, input_shape=(X_train.shape[1], 1)))
    model.add(Dropout(dropout_fraction))
    # Layer 2
    model.add(LSTM(units=number_units, return_sequences=True))
    model.add(Dropout(dropout_fraction))
    # Layer 3
    model.add(LSTM(units=number_units))
    model.add(Dropout(dropout_fraction))
    # Output layer
    model.add(Dense(1))
    
    # Compile the model
    model.compile(optimizer="adam", loss="mean_squared_error")
    
    # Optional: Summarize the model
    # model.summary()
    
    # Train the model
    print('Trainging the Model...')
    model.fit(X_train, y_train, epochs=5, shuffle=False, batch_size=1, verbose=1)
    
    # Evaluate the Model
    print(f"\n Loss: {model.evaluate(X_test, y_test)} \n")
    
    # Make some predictions
    predicted = model.predict(X_test)
    
    # Recover the original prices instead of the scaled version
    predicted_prices = scaler.inverse_transform(predicted)
    real_prices = scaler.inverse_transform(y_test.reshape(-1, 1))
    
    # Create a DataFrame of Real and Predicted values
    global predictions_df
    predictions_df = pd.DataFrame({"Real": real_prices.ravel(), "Predicted": predicted_prices.ravel()}, index = dataframe.index[-len(real_prices): ])
    
    
    # Plot the real vs predicted prices as a line chart
    # predictions_df.plot()
        
    return predictions_df.plot(), f'\n To view the dataframe, call predictions_df'
       
```

```python
# dataframe = the dataframe that your x (feature) and y (target) columns are in
dataframe = ohlc_df

# feature_column = the column number from your dataframe that contains your feature (x) inputs
feature_column_number = 3

# feature_column = the column number from your dataframe that contains your y (target) inputs
target_column_number = 3

# window_size = the number of rows the function will lookback and use to make a prediction (ie. minutes, days, months, etc.)
window_size = 1

# number of 
number_units = 5

# percent of data dropped from model to help prevent overfitting
dropout_fraction = 0.2

lstm_function(dataframe, feature_column_number, target_column_number, window_size, number_units, dropout_fraction)
```
